Risk Managment with ROAM report

Resolved: Risk has been eliminated and will not be a problem
Owned: Team member will be given ownership over a risk and entrusted to handle it
Accepted: It has been agreed that nothing will be done about it
Mitigated: Some action has been taken so tha the risk has been mitigated

You can creat a ROAM board so that you can create post-it notes with risks on them and then locate them physically in the catagory they should be placed into.

Escalating on a project is important to help manage conflict between parties.
Escalate and escalate early when there is the first sign of critical problems with time, project, or budget.

Escalation email:
1. Friendly tone
2. State your connection to the project
3. Explain the problem
4. Explain the consequences.
5. Make a request and ask for solutions

General principles of email writing
6. State what you want
7. Keep content concise
8. Structure your writing
9. Check grammar, punctuation, and spelling


Effective meeting principles
Try to be Collaborative and Inclusive if you can
