Four phases of software maangement:

Initiation:
Gather requirements and see if the project aligns with business needs
Outline project requirements and ifendity stakeholders and their roles

Planning:
Take business needs and make sure they are properly reflected in the work that is being done.
Also developing contingency plans to handle blocks in the execution phase.
Break down the project into the component parts and visuals

Execution:
Execute the project, monitor the progress and revise project plan when necsssary

Closure:
Review Performance

SMART Criteria of Project Goal:
Specific: Clearly Define what you want to accomplish
Measurable: Measure your metrics to track your progress and know when you have succeeded.
Attainable/Acheeivable: Objective is reasonable accomplished, scale project down?
Relevant: Ensure that goal aligns with broader goals and objectives
Timeframe: When you will start and stop the project
