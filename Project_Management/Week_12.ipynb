Risk Managment with ROAM report

Resolved: Risk has been eliminated and will not be a problem
Owned: Team member will be given ownership over a risk and entrusted to handle it
Accepted: It has been agreed that nothing will be done about it
Mitigated: Some action has been taken so tha the risk has been mitigated

## Agile and SCRUM ##

Agile vs Waterfall

# Waterfall #
Requirements: Formally approved project plans
Documentation: Many big and formal documents with rigorous approval processes
Deliverables: No deliverable released until the very end

# Agile # 
Requirements: Dynamic requirements that change as team receives feedback
Documentation: Short documents written as needed
Deliverables: Smaller and more frequent releases

Agile Manifesto:
Focuses on 
Individuals and Interactions
Deliver Working Software
Customer Collboration
Responding to Change

Problems with it?
 - Lack of clear goals sometimes
 - Difficulty predicting cost, time, or resources
 - Lack of executive sponsorship
 - Difficulty maintaining a level of collaboration
 - Putting process over behavior
 - Documentationcan be REALLY nice sometimes (it can cause a myriad of problems if not done properly)


Development Operations (DEVOPS)
Enables teams ot build, test, and release software faster and more reliably by incorporating agile principles and practices and improved collaboration between the dev team (who write application software) and the ops team (who deploys and manages the software).

DEVSECOPS is where you move security and operations as far left as possible.

# SCRUM #
SCRUM is a framework that materializes the agile philosophy
It helps team adopt agile

SCRUM requires a SCRUM Master (often the PM) to foster a environment where:
1. The prodect owner orders work for a complex problem into a product backlog
2. The scrum team turns a selection of the work into an increment of value during the sprint
3. The scrum team and its stakeholder inspect the result and adjust for the next sprint

SCRUM Pillars
 - Transparency: Make significant aspects of our work visible for those responsible for outcome
 - Inspection: Conduct timely checks towards the outcome of a sprint goal to detect undesirable variances
 - Adaptation: Search for ways to adjust the project, process, or product to minimize any further deviation or issues. (change what is not working)

SCRUM Values:
Commitment to acheiving goals
Courage to do the right thing (even if it is hard)
Focus means that team members are focusing on the tasks in front of them
Openness to show issues they are running into
Respect the opinions, skills, and independence of each other

SCRUM Roles:
Development Team Member
Product Owner
SCRUM Master
