## INITIATION PHASE ##
RACI Chart

RACI chart is used to clarify the roles of stakeholders in tasks

RACI Stands for:
Responsible: Do the work
Accountable: Delegate and review work
Consulted: Provide input and feedback on the work
Informed: Be looped into the progress

Each task should
At least one responsible person
Only one accountable person
One Consulted party per affected team (if any)

RACI may not be helpful with:
Certain project management methodologies (e.g. scrum)
Simple projects

Stakeholders in a RACI chart will be assigned different roles

## PLANNING PHASE ##
Breaking down a project into smaller steps
Create a budget
Risk management
Gantt Charts are good at breaking down tasks
Each of the major steps can be broken down into additional steps

Gantt Chart Planning
1. Idenfity tasks and subtasks
2. Identify taks connections
    a. Parallel tasks
    b. Sequential tasks
3. Create a timeline
4. Order tasks
5. Build the Gantt Chart
6. Assign tasks to team members
7. Monitor the project and update the chart as needed.

You an use LucidCharts to help create Gantt charts.

A PERT Chart:
You highlight critical paths to highlight the path that takes the longest to implement.
It focuses on execution threads and makes sure that managers are focusing on the correct execution thread to reduce the chance of delays.
A PERT chart is really only there to identify areas where the project could encounter chokeholds that could halt the project.

PERT charts are good for external users
Gantt charts are good for internal users
